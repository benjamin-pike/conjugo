```python
import os
import json
```

```python
top_dir = os.path.abspath('').replace('src', '')
out_dir = os.path.join(top_dir, 'out')
```

```python
audio = []

for language in ['spanish', 'french', 'german', 'italian', 'portuguese']:
    with open(os.path.join(out_dir, f'audio_{language}.json'), 'r', encoding =  'utf8') as file:
        audio_loaded = json.loads(file.read())

    rank = 1

    for verb in audio_loaded:
        audio.append(
            {
                '_id': f'{language}_{verb}',
                'language': language,
                'verb': verb,
                'rank': rank,
                'audio': audio_loaded[verb]
            }
        )

        rank += 1

with open(os.path.join(out_dir, f'audio_array.json'), 'w', encoding = 'utf8') as file:
    json.dump(audio, file, indent = 4, ensure_ascii = False)
```
