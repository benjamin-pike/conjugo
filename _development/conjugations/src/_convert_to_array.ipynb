```python
import os
import json
```

```python
top_dir = os.path.abspath('').replace('src', '')
data_dir = os.path.join(top_dir, 'data', 'language-specific')
out_dir = os.path.join(top_dir, 'out')
```

```python
conjugations = []

for language in ['spanish', 'french', 'german', 'italian', 'portuguese']:
    with open(os.path.join(data_dir, f'conjugations_{language}.json'), 'r', encoding =  'utf8') as file:
        conjugations_loaded = json.loads(file.read())

    rank = 1

    for verb in conjugations_loaded:
        conjugations.append(
            {
                '_id': f'{language}_{verb}',
                'language': language,
                'verb': verb,
                'rank': rank,
                'conjugations': conjugations_loaded[verb]
            }
        )

        rank += 1

with open(os.path.join(out_dir, f'conjugations_array.json'), 'w', encoding = 'utf8') as file:
    json.dump(conjugations, file, indent = 4, ensure_ascii = False)
```
