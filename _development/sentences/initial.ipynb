```python
import os
import json
from collections import Counter
```

```python
path_conjugations = os.path.join(os.pardir, 'conjugations', 'data', 'language-specific')
```

```python
language = 'spanish'
```

```python
with open(os.path.join(path_conjugations, f'conjugations_{language}.json'), 'r', encoding = 'utf8') as file:
    raw = json.loads(file.read())
    conjugations = {verb: raw[verb] for verb in raw if raw[verb]['rank'] <= 2000}
```

```python
all_conjugations = []

for verb in conjugations:
    for complexity in ['simple', 'compound', 'progressive']:
        for mood in conjugations[verb][complexity]:
            for tense in conjugations[verb][complexity][mood]:
                for subject in ["yo", "tu", "el", "nosotros", "vosotros", "ellos"]:
                    try:
                        conjugation = conjugations[verb][complexity][mood][tense][subject]

                        all_conjugations.append( ( conjugation, [verb, complexity, mood, tense, subject] ))    
                    except KeyError:
                        pass
                        
```

```python
sorted_conjugations = sorted( list( dict( Counter( [ _[0] for _ in all_conjugations ] )).items() ), key = lambda x: x[1], reverse = True )
```

```python
duplicates = [ x for x in sorted_conjugations if x[1] > 1 ]
```

```python
duplicate_conjugations = []

for duplicate in duplicates:
    conjugation = duplicate[0]

    for c in all_conjugations:
        if c[0] == conjugation:
            duplicate_conjugations.append(c)
```
