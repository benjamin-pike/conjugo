```python
import json
```

```python
with open("./conjugations/data/language-specific/conjugations_italian.json", 'r', encoding = 'utf-8') as file:
    conjugations = json.loads( file.read() )
```

```python
abnormal = set()

for verb in conjugations:
    for complexity in [ "simple", "compound" ]:
        for mood in conjugations[verb][complexity]:
            if mood != "imperative":
                for tense in conjugations[verb][complexity][mood]:
                    rank = conjugations[verb]["rank"]
                    lui = conjugations[verb][complexity][mood][tense]['lui'] 
                    lei = conjugations[verb][complexity][mood][tense]['lei'] 

                    if lui != lei:
                        # abnormal.add( f'{complexity}-{mood}-{tense}' )
                        print(lui, lei)
```
