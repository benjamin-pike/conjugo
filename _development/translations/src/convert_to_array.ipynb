```python
import os
import json
```

```python
top_dir = os.path.abspath('').replace('src', '')
out_dir = os.path.join(top_dir, 'out')
```

```python
translations = []

for language in ['spanish', 'french', 'german', 'italian', 'portuguese']:
    with open(os.path.join(out_dir, f'translations_{language}.json'), 'r', encoding =  'utf8') as file:
        translations_loaded = json.loads(file.read())

    rank = 1

    for verb in translations_loaded:
        translations.append(
            {
                '_id': f'{language}_{verb}',
                'language': language,
                'verb': verb,
                'rank': rank,
                'translations': translations_loaded[verb]
            }
        )

        rank += 1

with open(os.path.join(out_dir, f'translations_array.json'), 'w', encoding = 'utf8') as file:
    json.dump(translations, file, indent = 4, ensure_ascii = False)
```
