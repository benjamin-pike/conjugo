```python
import os
import json
from pathlib import Path
from profanity_filter import ProfanityFilter
```

```python
language = 'portuguese'
pf = ProfanityFilter()
```

```python
path = os.path.abspath('').replace('src', '')
top_dir = Path(path).parent.absolute()
conjugations_data_dir = os.path.join(top_dir, 'conjugations', 'data', 'language-specific')
out_dir = os.path.join(path, 'out')
```

```python
with open(os.path.join(out_dir, f'translations_{language}.json'), 'r', encoding = 'utf8') as file:
    translations = json.loads(file.read())

with open(os.path.join(conjugations_data_dir, f'conjugations_{language}.json'), 'r', encoding = 'utf8') as file:
    conjugations = json.loads(file.read())
```

```python
for verb in translations:
    pre = translations[verb]
    post = []

    for translation in pre:
        if '*' not in pf.censor(translation):
            post.append(translation)

    translations[verb] = post
```

```python
for verb in translations:
    remove = []
    
    for i in range(0, len(translations[verb])):
        if translations[verb][i].strip()[:3] == 'to ':
            candidate = translations[verb][i].strip()[3:]
            
            if candidate not in translations[verb]:
                translations[verb][i] = candidate
            else:
                remove.append(translations[verb][i])

    for translation in remove:
        translations[verb].remove(translation)
```

```python
for verb in translations:    
    for i in range(0, len(translations[verb])):
        translations[verb][i] = translations[verb][i].replace(' sth', ' something').replace(' sb', ' somebody')
```

```python
for verb in translations:
    remove = []

    for translation in translations[verb]:
        if len(translation.split(' ')) > 5 or len(translation) > 20:
            remove.append(translation)

    for translation in remove:
        translations[verb].remove(translation)
```

```python
for verb,rank in zip(translations, range(1, len([verb for verb in translations]) + 1)):
    if rank <= 1000 and translations[verb] == []:
        print(verb, rank)
```

```python
for verb in additions[language]:
    translations[verb] = additions[language][verb]
```

```python
empty = [verb for verb in translations if translations[verb] == []]

for verb in empty:
    translations.pop(verb)
    conjugations.pop(verb)
```

```python
if [verb for verb in translations][1999] == [verb for verb in conjugations][1999]:
    rank = 1

    for verb in conjugations:
        conjugations[verb]['rank'] = rank
        rank += 1
```

```python
new = {}

for verb in translations:
    if len(new) < 2000:
        new[verb] = translations[verb]
```

```python
with open(os.path.join(out_dir, f'translations_{language}.json'), 'w', encoding = 'utf8') as file:
    json.dump(new, file, indent = 4, ensure_ascii = False)

with open(os.path.join(conjugations_data_dir, f'conjugations_{language}.json'), 'w', encoding = 'utf8') as file:
    json.dump(conjugations, file, indent = 4, ensure_ascii = False)
```
